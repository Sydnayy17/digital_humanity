````python

import re

def extract_year(date_str):
    """ Extracts the year from different date formats """
    if pd.isna(date_str):  # Handle NaN values
        return None

    # Handle MM/DD/YY format, e.g., '02/28/66'
    match = re.search(r'\d{2}/\d{2}/(\d{2})$', date_str)
    if match:
        year = int(match.group(1))
        return 2000 + year if year < 24 else 1900 + year  # Assume 00-23 corresponds to 2000s, 24-99 to 1900s

    # Handle "Month Day, YYYY" format, e.g., 'January 1st 2011'
    match = re.search(r'(\d{4})', date_str)
    if match:
        return int(match.group(1))

    return None  # Ignore other formats

# 應用函數到 DataFrame
book_pd["year"] = book_pd["publishDate"].apply(extract_year)
book_pd
````

## 2-1. Basic Code for Plotting Trees in Python

[NetworkX](https://networkx.org/) is a powerful Python library for creating, manipulating, and analyzing complex networks and graph structures. It provides an easy-to-use interface for working with nodes and edges, enabling users to model relationships such as social networks, transportation systems, character interactions in books, and hierarchical structures. With support for graph algorithms like shortest paths, centrality measures, and community detection, NetworkX is widely used in data science, machine learning, and network analysis. It integrates seamlessly with Matplotlib for visualization, allowing users to explore and interpret graph structures interactively. Whether you are analyzing small-scale trees or large real-world networks, NetworkX provides the necessary tools for efficient and scalable graph analysis.

````python
import networkx as nx
````

````python
G=nx.Graph()

#what will happen if we dont specifies the nodes first?
G.add_edge("A","B")

nx.draw(G, with_labels=True)
````

````python
#weighted Network
G=nx.Graph()

G.add_edge("A","B",weight=60)
G.add_edge("A","D",weight=3)
G.add_edge("A","C",weight=0.1)
G.add_edge("B","D",weight=1)

# draw the nodes with labels
nx.draw(G, with_labels=True)
````

````python
#directed graph
G_Gdi=nx.DiGraph()

G.add_edge("A","B",weight=1)
G.add_edge("A","D",weight=3)
G.add_edge("A","C",weight=1)
G.add_edge("B","D",weight=2)

# create a dictionary of positions for the nodes
pos = nx.spring_layout(G)

# create a dictionary of edge labels
edge_labels = {(u, v): d['weight'] for u, v, d in G.edges(data=True)}

# draw the edge labels
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
nx.draw(G, pos, with_labels=True)
````

````python

G=nx.MultiGraph()

G.add_edge("A","B",weight=1,relation="family")
G.add_edge("A","D",weight=3,relation="family")
G.add_edge("A","C",weight=1,relation="Work")
G.add_edge("B","D",weight=2,relation="Friend")

# create a dictionary of positions for the nodes
pos = nx.spring_layout(G)

# create a dictionary of edge labels
edge_labels = {(u, v): d['relation'] for u, v, d in G.edges(data=True)}

# draw the edge labels
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

nx.draw(G, pos, with_labels=True)
````
